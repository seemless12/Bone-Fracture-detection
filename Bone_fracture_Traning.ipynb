```python
from ultralytics import YOLO
```

```python
Model = YOLO("yolov8n.pt")
```

```python
yaml_path = "/content/bone fracture/Human Bone Fractures Multi-modal Image Dataset (HBFMID)/Bone Fractures Detection/data.yaml"
```

```python
Model.train (
    data=yaml_path,
    epochs=100,
    imgsz=640,
    batch=16,
    patience=20,
    pretrained=True
)
```
